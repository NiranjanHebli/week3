```python
import pandas as pd
import numpy as np
```

```python
def create_sample_data():
    """Create 25 realistic e-commerce products"""
    np.random.seed(42)
    categories = np.random.choice(['Electronics', 'Clothing', 'Books', 'Home'], size=25)
    names = [f'Product_{i}' for i in range(25)]
    prices = np.random.uniform(100, 20000, 25).round(2)
    stocks = np.random.randint(0, 500, 25)
    ratings = np.random.uniform(1, 5, 25).round(1)
    reviews = np.random.randint(0, 1000, 25)

    return pd.DataFrame({
        'name': names,
        'category': categories,
        'price': prices,
        'stock': stocks,
        'rating': ratings,
        'num_reviews': reviews
    })
```

```python
df = create_sample_data()
```

```python
def first_five_minutes_checklist(df):
    print("=== First 5 Minutes Checklist ===\n")
    print(f"Shape: {df.shape}")
    print(f"\nHead:\n{df.head()}")
    print("\nInfo:")
    df.info()
    print(f"\nDescribe:\n{df.describe()}")
    print(f"\nDtypes:\n{df.dtypes}")
    print(f"\nColumns: {df.columns.tolist()}")
    print(f"\nIndex: {df.index}")
    print(f"\nMissing values:\n{df.isnull().sum()}")
    print(f"Memory usage: {df.memory_usage(deep=True).sum()} bytes\n")
```

```python
first_five_minutes_checklist(df)
```

```python

def demonstrate_loc(df):
    """Demonstrate .loc operations"""
    print("=== .loc Examples ===\n")
    
    # (a) All Electronics
    electronics = df.loc[df['category'] == 'Electronics']
    print("All Electronics:\n", electronics[['name', 'price', 'rating']])
    
    # (b) Rating >4.0 and price <5000
    high_rated_cheap = df.loc[(df['rating'] > 4.0) & (df['price'] < 5000)]
    print("\nRated >4.0 and price <5000:\n", high_rated_cheap[['name', 'price', 'rating']])
    
    # (c) Update stock for Product_0
    original_stock = df.loc[df['name'] == 'Product_0', 'stock'].values[0]
    df.loc[df['name'] == 'Product_0', 'stock'] = 999
    print(f"\nUpdated Product_0 stock: {original_stock} -> 999")
```

```python
demonstrate_loc(df)
```

```python
def demonstrate_iloc(df):
    """Demonstrate .iloc operations"""
    print("\n=== .iloc Examples ===\n")
    
    # (a) First 5 and last 5 products
    print("First 5:\n", df.iloc[:5][['name', 'category', 'price']])
    print("\nLast 5:\n", df.iloc[-5:][['name', 'category', 'price']])
    
    # (b) Every other row
    print("\nEvery other row (first 6):\n", df.iloc[::2].head()[['name', 'price']])
    
    # (c) Rows 10-15, columns 0-3
    print("\nRows 10-15, cols 0-3:\n", df.iloc[10:16, 0:4])
```

```python
demonstrate_iloc(df)
```

```python
def create_filtered_dataframes(df):
    """Create 3 filtered DataFrames"""
    budget_products = df[df['price'] < 1000].copy()
    premium_products = df[df['price'] > 10000].copy()
    popular_products = df[(df['num_reviews'] > 100) & (df['rating'] > 4.0)].copy()
    
    print("\n=== Filtered DataFrames ===\n")
    print(f"Budget products (price < 1000): {len(budget_products)} rows")
    print(f"Premium products (price > 10000): {len(premium_products)} rows")
    print(f"Popular products (reviews >100 & rating >4.0): {len(popular_products)} rows")
    
    return {
        'budget_products': budget_products,
        'premium_products': premium_products,
        'popular_products': popular_products
    }
```

```python
filtered_dfs = create_filtered_dataframes(df)
```

```python

def export_to_csv(filtered_dfs):
    """Export filtered DataFrames to separate CSV files using a loop"""
    print("\n=== Exporting CSVs ===\n")
    for name, dfilt in filtered_dfs.items():
        filename = f"{name}.csv"
        dfilt.to_csv(filename, index=False)
        print(f"- Exported {filename} ({len(dfilt)} rows)")
    print("\nAnalysis complete! Check generated CSV files.")
    print("\nSummary:")
    print(f" - Total products: {len(df)}")
    print(f" - Budget: {len(filtered_dfs['budget_products'])}")
    print(f" - Premium: {len(filtered_dfs['premium_products'])}")
    print(f" - Popular: {len(filtered_dfs['popular_products'])}")
```

```python
export_to_csv(filtered_dfs)
```
