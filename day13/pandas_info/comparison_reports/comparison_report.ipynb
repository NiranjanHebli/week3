```python
import pandas as pd
import numpy as np
```

```python
def create_monthly_sales_data(month, n_orders=25):
    """Create sales DataFrame for a single month """
    np.random.seed(42 + ['Jan', 'Feb', 'Mar'].index(month))
    products = ['Laptop', 'Phone', 'Tablet', 'Headphones', 'Keyboard']
    
    df = pd.DataFrame({
        'Product': np.random.choice(products, n_orders),
        'Quantity': np.random.randint(1, 10, n_orders),
        'Price': np.random.uniform(50, 1000, n_orders)
    })
    df['Revenue'] = df['Quantity'] * df['Price']
    df['Order_Value'] = df['Revenue'] / df['Quantity']
    df['Month'] = month  # Add Month column
    
    return df
```

```python
def calculate_monthly_metrics(df):
    """Calculate key metrics for a month"""
    total_revenue = df['Revenue'].sum()
    avg_order_value = df['Revenue'].mean()
    
    # Top-selling product by total revenue
    top_product = df.loc[
        df.groupby('Product')['Revenue'].idxmax()
    ]['Product'].iloc[0]
    
    return {
        'Total_Revenue': total_revenue,
        'Avg_Order_Value': avg_order_value,
        'Top_Product': top_product
    }
```

```python
def find_outliers(df):
    """Find high/low revenue outliers using query, nlargest, nsmallest"""
    high_revenue = df.query('Revenue > 2000')
    top_orders = df.nlargest(3, 'Revenue')[['Product', 'Quantity', 'Revenue', 'Month']]
    low_orders = df.nsmallest(3, 'Revenue')[['Product', 'Quantity', 'Revenue', 'Month']]
    
    return {
        'high_revenue_orders': high_revenue[['Product', 'Quantity', 'Revenue', 'Month']],
        'top_3_orders': top_orders,
        'bottom_3_orders': low_orders
    }
```

```python
# Generate 3 months of data
months = ['Jan', 'Feb', 'Mar']
dfs = {month: create_monthly_sales_data(month) for month in months}

# Display first few rows
for month, df in dfs.items():
    print(f"{month} shape: {df.shape}")
    print(df.head())
    print()
```

```python
metrics = {}
for month in months:
    month_metrics = calculate_monthly_metrics(dfs[month])
    month_metrics['Month'] = month  # Add Month column to metrics
    metrics[month] = month_metrics

# Create summary with Month as first column
summary_data = []
for month, data in metrics.items():
    row = {'Month': month, **data}
    summary_data.append(row)

summary_df = pd.DataFrame(summary_data).round(2)
summary_df = summary_df[['Month', 'Total_Revenue', 'Avg_Order_Value', 'Top_Product']]  # Reorder columns

print("Summary Comparison Table:")
print(summary_df)
```

```python
for month in months:
    print(f"\n{'='*35}")
    print(f"       { month.upper()} OUTLIERS ANALYSIS")
    print(f"{'='*35}")
    
    outliers = find_outliers(dfs[month])
    
    print("\nHigh Revenue Orders (>2000):")
    print(outliers['high_revenue_orders'])
    
    print("\nTop 3 Orders:")
    print(outliers['top_3_orders'])
    
    print("\nBottom 3 Orders:")
    print(outliers['bottom_3_orders'])
```

```python
# Save summary to CSV
summary_df.to_csv('sales_comparison_summary.csv', index=False)

# Save combined detailed data
combined_data = pd.concat(dfs.values(), ignore_index=True)
combined_data.to_csv('monthly_sales_data.csv', index=False)

print("Files saved:")
print("- sales_comparison_summary.csv")
print("- monthly_sales_data.csv ")
print("\nSummary Preview:")
print(summary_df)
```
