```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Simulate the Central Limit Theorem for a given distribution
def simulate_clt(distribution, params, n_samples, n_simulations):
    """Simulate the Central Limit Theorem for a given distribution."""

    # Generate sample means
    sample_means = []

    for _ in range(n_simulations):
        sample = distribution.rvs(*params, size=n_samples)
        sample_means.append(np.mean(sample))

    sample_means = np.array(sample_means)

    # Theoretical mean and variance of original distribution
    mu = distribution.mean(*params)
    sigma = distribution.std(*params)

    # CLT normal approximation
    clt_mean = mu
    clt_std = sigma / np.sqrt(n_samples)

    # Histogram of simulated sample means
    plt.hist(sample_means, bins=30, density=True, alpha=0.6)

    # Overlay theoretical normal approximation
    x = np.linspace(sample_means.min(), sample_means.max(), 500)
    plt.plot(x, norm.pdf(x, clt_mean, clt_std))

    # Labels
    plt.title(f'CLT Simulation: n={n_samples}, simulations={n_simulations}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')

    plt.show()
```

```python
# Example usage with a uniform distribution
from scipy.stats import uniform
simulate_clt(uniform, (0, 1), n_samples=30, n_simulations=1000)
```

```python

# Example usage with a normal distribution
from scipy.stats import norm
simulate_clt(norm, (0, 1), n_samples=30, n_simulations=1000)
```
