
## 1) Compare `np.random.default_rng()` vs `np.random.seed()`

```python
import numpy as np
import time

# Legacy API
start1 = time.time()
np.random.seed(2026)
legacy_samples = np.random.normal(0, 1, 1_000_000)
end1 = time.time()

# New API
start2 = time.time()
rng = np.random.default_rng(2026)
new_samples = rng.normal(0, 1, 1_000_000)
end2 = time.time()

print("Legacy API time:", end1 - start1)
print("default_rng() time:", end2 - start2)
```

## 2) Synthetic dataset for linear regression

```python
import numpy as np

def generate_linear_data(seed=2026):
    rng = np.random.default_rng(seed)

    X = rng.normal(0, 1, (100, 3))
    w = np.array([2.5, -1.3, 0.7])
    noise = rng.normal(0, 0.5, 100)

    y = X @ w + noise
    return X, y, w

X, y, w = generate_linear_data()

print("X shape:", X.shape)
print("First 5 y values:", y[:5])
print("True weights:", w)
```


## 3) What is np.random.Generator and why introduced?

np.random.Generator is the modern random-number interface introduced to replace the older global random API. It separates the random-number engine from sampling methods, making results more reproducible and safer in multi-module code because each generator instance is independent. It also uses better underlying algorithms such as **PCG64**, which provide improved statistical quality and performance. The older np.random.seed() modifies global state, while Generator avoids unintended interference between experiments.
