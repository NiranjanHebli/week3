```python
import numpy as np
```

### 1. Create 200x8 random grade matrix

```python
np.random.seed(2026)
grades = np.random.randint(0, 101, size=(200, 8))
```

### 2. Calculate statistics

```python
student_avg = grades.mean(axis=1)      # per-student average
course_avg = grades.mean(axis=0) 
```

```python
overall_mean = grades.mean()
overall_std = grades.std()
overall_min = grades.min()
overall_max = grades.max()
```

### 3. Apply curve: courses with avg < 50 get +10 points, cap at 100

```python

curve_mask = course_avg < 50
grades_curved = np.minimum(grades + curve_mask * 10, 100)
```

 ### 4. Assign letter grades using np.where chains

```python
letter_grades = np.where(grades_curved >= 90, 'A',
                np.where(grades_curved >= 80, 'B',
                np.where(grades_curved >= 70, 'C',
                np.where(grades_curved >= 60, 'D', 'F'))))
```

###  5. Top 10 students by overall average

```python
curved_student_avg = grades_curved.mean(axis=1)
top10_students = np.argsort(curved_student_avg)[-10:][::-1]
```

### 6. Students who passed ALL courses

```python
passed_all = (grades_curved >= 60).all(axis=1)
```

```python
num_passed = passed_all.sum()
```

```python
print("Per-student average:", student_avg)
print("Per-course average:", course_avg)
print("Overall mean/std/min/max:", overall_mean, overall_std, overall_min, overall_max)
print("Top 10 student indices:", top10_students)
print("Passed all courses:", passed_all)
print(f"Students passing all courses: {num_passed} out of 200")
```
