```python
import numpy as np
```

### 1. Generate dataset: (50 sensors, 24 hours, 3 metrics)

```python

np.random.seed(1313)
```

```python
temperature = np.random.randint(15, 46, size=(50, 24, 1))   # 15-45
humidity = np.random.randint(20, 96, size=(50, 24, 1))      # 20-95
battery = np.random.randint(10, 101, size=(50, 24, 1))      # 10-100
```

```python

data = np.concatenate([temperature, humidity, battery], axis=2)
```

### 2. Alert sensors: temp > 40 OR humidity > 90 in any hour

```python
alert_mask = ((data[:, :, 0] > 40) | (data[:, :, 1] > 90)).any(axis=1)
alert_sensors = np.where(alert_mask)[0]
```

### 3. Per-sensor daily averages (50 x 3)

```python
daily_avg = data.mean(axis=1)
```

### 4. Hour with highest average temperature across all sensors

```python
hourly_temp_avg = data[:, :, 0].mean(axis=0)
hottest_hour = np.argmax(hourly_temp_avg)
```

### 5. Battery drain analysis

```python
battery_drop = data[:, 0, 2] - data[:, -1, 2]
critical_battery = np.where(battery_drop > 50)[0]
```

### 6. Normalize all metrics to [0,1] using min-max normalization

```python

metric_min = data.min(axis=(0, 1), keepdims=True)
metric_max = data.max(axis=(0, 1), keepdims=True)

normalized = (data - metric_min) / (metric_max - metric_min)
```

### 7. Save daily averages to CSV

```python
np.savetxt(
    "sensor_summary.csv",
    daily_avg,
    delimiter=",",
    header="temperature_avg,humidity_avg,battery_avg",
    comments=""
)
```

```python

# Expected output format
print("Alert sensors:", alert_sensors.tolist())
print("Hottest hour:", hottest_hour)
print("Sensors with critical battery drain:", critical_battery.tolist())
print("sensor_summary.csv:", daily_avg.shape[0], "rows x", daily_avg.shape[1], "columns")
```
