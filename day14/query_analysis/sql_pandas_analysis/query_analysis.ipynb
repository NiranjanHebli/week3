###  7. Get all sales reps with their customer count (including those with zero customers)

```python
sales_rep_customers = employees.merge(
    customers,
    left_on='employeeNumber',
    right_on='salesRepEmployeeNumber',
    how='left'
)

sales_rep_customer_count = (
    sales_rep_customers[sales_rep_customers['jobTitle'] == 'Sales Rep']
    .groupby(['employeeNumber', 'firstName', 'lastName'])
    .agg(customers=('customerNumber', 'count'))
    .reset_index()
)

sales_rep_customer_count
```

### 8. Get all managers with their direct reports count

```python
manager_reports = employees.merge(
    employees,
    left_on='employeeNumber',
    right_on='reportsTo',
    suffixes=('_manager', '_employee')
)

manager_reports = manager_reports[
    manager_reports['jobTitle_manager'].str.contains('Manager', na=False)
]

manager_reports[['firstName_manager', 'lastName_manager', 'firstName_employee', 'lastName_employee']]
```

### 9. Get all sales reps who do not have any customers assigned

```python
sales_reps_no_customers = sales_rep_customers[
    (sales_rep_customers['jobTitle'] == 'Sales Rep') &
    (sales_rep_customers['customerNumber'].isna())
]

sales_reps_no_customers
```

### 10. Get the number of sales reps per country

```python
sales_reps_per_country = (
    emp_office[emp_office['jobTitle'] == 'Sales Rep']
    .groupby('country')
    .size()
    .reset_index(name='reps')
)

sales_reps_per_country
```

### 11. Get the average salary per office

```python
avg_salary_per_office = (
    employees.groupby('officeCode')['salary']
    .mean()
    .reset_index(name='avg_salary')
)

avg_salary_per_office
```

###  12. Get the number of employees per country

```python
employees_per_country = (
    emp_office.groupby('country')
    .size()
    .reset_index(name='no_of_employees')
)

employees_per_country
```

### 13. Get the job title with the highest number of employees

```python
top_job_title = employees['jobTitle'].value_counts().reset_index().head(1)
top_job_title
```

### 14. Get the maximum salary for each manager (reportsTo)

```python
max_salary_per_manager = (
    employees.dropna(subset=['reportsTo'])
    .groupby('reportsTo')['salary']
    .max()
    .reset_index(name='max_salary')
)

max_salary_per_manager
```

### 15. Get the number of employees reporting to VPs

```python
vp_ids = employees[
    employees['jobTitle'].str.contains('VP', na=False)
]['employeeNumber']

reports_to_vps = (
    employees[employees['reportsTo'].isin(vp_ids)]
    .groupby('jobTitle')
    .size()
    .reset_index(name='count')
)

reports_to_vps
```

### 16. Get the ranking of sales reps based on their customer count, partitioned by country and globally

```python
rank_df = (
    sales_rep_customers[sales_rep_customers['jobTitle'] == 'Sales Rep']
    .groupby(['employeeNumber', 'firstName', 'lastName', 'officeCode'])
    .agg(customer_count=('customerNumber', 'count'))
    .reset_index()
)

rank_df = rank_df.merge(offices[['officeCode', 'country']], on='officeCode')

rank_df['country_rank'] = rank_df.groupby('country')['customer_count'].rank(method='dense', ascending=False)
rank_df['global_rank'] = rank_df['customer_count'].rank(method='dense', ascending=False)

rank_df
```

### 17. Get the top 3 highest paid employees in each office

```python
top_paid = employees.copy()

top_paid['rank'] = top_paid.groupby('officeCode')['salary'].rank(method='dense', ascending=False)

top3_paid = top_paid[top_paid['rank'] <= 3]

top3_paid
```


### 18. Get the month-over-month salary growth percentage for each employee

```python
monthly_salaries = monthly_salaries.sort_values(['employeeNumber', 'month'])

monthly_salaries['prev_month'] = monthly_salaries.groupby('employeeNumber')['salary'].shift(1)

monthly_salaries['growth_pct'] = (
    (monthly_salaries['salary'] - monthly_salaries['prev_month']) /
    monthly_salaries['prev_month'] * 100
)
monthly_salaries
```

### 19. Get all employees whose salary is above the average salary of their office

```python
overall_avg = employees['salary'].mean()

high_earners = employees[employees['salary'] > overall_avg]

office_avg = employees.groupby('officeCode')['salary'].mean().reset_index(name='avg')

high_earners_above_avg = high_earners.merge(office_avg, on='officeCode')

high_earners_above_avg = high_earners_above_avg[
    high_earners_above_avg['salary'] > high_earners_above_avg['avg']
]

high_earners_above_avg
```

### 20. Get all employees whose salary is above the average salary of their office (without using CTEs)

```python
office_mean = employees.groupby('officeCode')['salary'].transform('mean')

salary_above_office_avg = employees[employees['salary'] > office_mean]

print(salary_above_office_avg)
```

### Report one insight about how PostgreSQL executes each.

#### Filter Query (WHERE salary > 100000)
- PostgreSQL usually performs a Sequential Scan, meaning it checks each row because the table is small or no index exists on salary.

#### Join Query (employees JOIN offices)
- PostgreSQL often uses a Hash Join, where it creates a hash table for the smaller table (offices) to match rows efficiently.

#### Aggregation Query (GROUP BY officeCode)
- PostgreSQL typically uses HashAggregate, which groups rows in memory before calculating aggregate values like averages.
